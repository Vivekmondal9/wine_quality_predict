```python
# Importing all necessery libraries.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fetching the dataset using pandas library.

data=pd.read_csv("wine_quality.csv",header=0)
data.head()
```

```python
# Shape of the dataset
data[data.columns[0]]
```

<font size=5><u><b>We can see that the datasets has totall 1599 records and 12 columns in it.

```python
# Checking the data type of each columns of the dataset.
data.info()
```

<font size=5><u><b>It is visible that datatype of all the columns are "float"

```python
# Null value checking
((data.isnull().sum())/len(data))*100
```

```python
data.columns
```

<font size=5><u><b>We can see that the dataset has a very few % of null value.

```python
# Filling the Null value with the mean value of that column.
data["total sulfur dioxide"].fillna(data["total sulfur dioxide"].mean(),inplace=True)
data["pH"].fillna(data["pH"].mean(),inplace=True)
data["quality"].fillna(data["quality"].mean(),inplace=True)
```

```python
((data.isnull().sum())/len(data))*100
```

<font size=5><u><b>Now we can see that there is no Null value in the datasets.

```python
# Checking Data distribution in "fixed acidity".
data["fixed acidity"].describe()
```

```python
sns.boxplot(data["fixed acidity"])
plt.show()
```

<b>Most of the data are lying between 7 to 9

```python
# Analysing the quality with each column

plt.figure(figsize=(20,10))
sns.barplot(data=data,x="fixed acidity",y="quality")
plt.title("Quality vs Fixed acidity",fontdict={"fontsize":30,"fontweight":10,"color":"red"})
plt.xlabel("Fixed Acidity",fontdict={"fontsize":15,"fontweight":10,"color":"black"})
plt.ylabel("Quality",fontdict={"fontsize":15,"fontweight":10,"color":"black"})
plt.xticks(rotation=90)
plt.show()
```

<font size=4><i>From the above bar plot we can clearly see that quality is  dependent on the Fixed acidity.
